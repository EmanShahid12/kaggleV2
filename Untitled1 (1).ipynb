```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# ------------------------------
# Locate dataset
# ------------------------------
filename = "KaggleV2-May-2016.xlsx"
dataset_path = None

# Try fixed path first
test_path = r"C:\Users\BRU\Desktop\numpyy\KaggleV2-May-2016.xlsx"
if os.path.exists(test_path):
    dataset_path = test_path
else:
    print("Searching C:\\ drive for the file (this may take time)...")
    for root, dirs, files in os.walk("C:\\"):
        for file in files:
            if file.lower() == filename.lower():
                dataset_path = os.path.join(root, file)
                break
        if dataset_path:
            break

if not dataset_path:
    raise FileNotFoundError(f"Could not find '{filename}' anywhere on your system.")

print(f"✅ Found dataset at: {dataset_path}")

# ------------------------------
# Load dataset
# -----------------------------
df = pd.read_excel(dataset_path)
print("\n--- Dataset Shape ---")
print(df.shape)

print("\n--- Dataset Info ---")
df.info()

print("\n--- Missing Values ---")
print(df.isnull().sum())

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)

print("\n--- Full Dataset ---")
print(df)
df['ScheduledDay'] = pd.to_datetime(df['ScheduledDay'])
df['AppointmentDay'] = pd.to_datetime(df['AppointmentDay'])
df['waiting_days'] = (df['AppointmentDay'] - df['ScheduledDay']).dt.days
df['is_weekend'] = df['AppointmentDay'].dt.weekday >= 5
df['No-show'] = df['No-show'].map({'No': 0, 'Yes': 1})

print("\nFeature engineering completed.")
plt.figure(figsize=(8,5))
df['Age'].hist(bins=30, color='skyblue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
plt.figure(figsize=(8,5))
sns.countplot(x='No-show', data=df)
plt.title('No-show vs Show Count')
plt.show()

plt.figure(figsize=(10,6))
sns.boxplot(x='No-show', y='Age', data=df)
plt.title('Age Distribution by No-show Status')
plt.show()
fig1 = px.histogram(df, x='Age', color='No-show', barmode='overlay',
                    title="Age Distribution by No-show Status")
fig1.show()

fig2 = px.scatter(df, x='waiting_days', y='Age', color='No-show',
                  title="Waiting Days vs Age by No-show Status")
fig2.show()
neigh_counts = df['Neighbourhood'].value_counts().reset_index()
neigh_counts.columns = ['Neighbourhood', 'count']

fig3 = px.bar(neigh_counts,
              x='Neighbourhood', y='count',
              title="Appointments by Neighbourhood")
fig3.update_layout(xaxis_title="Neighbourhood", yaxis_title="Count", xaxis_tickangle=-90)
fig3.show()
save_path = os.path.join(os.path.expanduser("~"), "Desktop", "cleaned_kaggle_dataset.csv")
df.to_csv(save_path, index=False)
print(f" Cleaned dataset saved to: {save_path}")
```
